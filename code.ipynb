#### This code processes log files and the final output is a merged csv file of all evaluations (for now)

```python
from extract_log_data import process_logs
import pandas as pd
```

```python
df_prompts = pd.read_csv("data//prompts.csv", sep="|")
df_prompts.head()
```

```python
df_prompts.tail()
```

```python
source_loc = "data//logs"
destination_loc = "data//processed_logs"

df_merged = process_logs(source_loc, destination_loc)
```

```python
df_merged = df_merged.sort_values(by=["prompt_id", "config"])
df_merged
```

```python
col_keep = ["prompt_id", "user_prompt"]
df_final = pd.merge(df_merged, df_prompts[col_keep], on="prompt_id", how="left")
df_final.sort_values(by="prompt_id", inplace=True)
df_final.reset_index(inplace=True, drop=True)
df_final
```

```python
df_final.to_csv("data//evaluations.csv", index=False, sep="|")
```
